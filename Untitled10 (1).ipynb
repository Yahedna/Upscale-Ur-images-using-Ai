# DATA Preparation

to train our model we used this dataset from Kaggle
https://www.kaggle.com/datasets/hsankesara/flickr-image-dataset


First, the original images were loaded and converted to np.float32 format to ensure compatibility with the model input requirements.

From each original image, two resized versions were created:

A 128×128 version to serve as the low-resolution input.

A 256×256 version put into 'out_256' and 'out_256b', for 2x upscaling test.

A 384×384 version labeled 'out_384', for 3x upscaling test.


The final preprocessed dataset would look like <|input|>, <|Y1|Y2|Y1|>
Y1 and Y2 are 256x256 and 384x384 expected output "ytrue"




#MODEL  ARCH

```python
import tensorflow as tf
from tensorflow.keras import layers, models


def upscaling_model():
    inputs = tf.keras.Input(shape=(None, None, 3))
    x1 = layers.Conv2D(64, 3, padding='same', activation='relu')(inputs)
    x1 = layers.Conv2DTranspose(3, kernel_size=4, strides=2, padding='same')(x1)

    x2 = layers.Conv2D(128, 4, padding='same', activation='relu')(inputs)
    x2 = layers.Conv2DTranspose(3, kernel_size=3, strides=3, padding='same')(x2)

    x3 = layers.Conv2D(256, 2, padding='same', activation='relu')(inputs)
    x3 = layers.Conv2DTranspose(3, kernel_size=2, strides=2, padding='same')(x3)
    # Average all
    return tf.keras.Model(inputs=inputs, outputs={
    'out_256': x1,  # 2X
    'out_384': x2,  # 3X
    'out_256_b': x3 # 2X (NON recursive )
})




#model = upscaling_model()
model = tf.keras.models.load_model('/content/Mymodel.keras')



model.summary()

```

# PREDICTION


```python

orignalimage='/content/Urban46.png'
image_raw = tf.io.read_file(orignalimage)
image = tf.image.decode_png(image_raw)
```

```python
import numpy as np
from PIL import Image

# The model expects a batch of images, so we add a batch dimension.
input_image = tf.expand_dims(image, axis=0)

# Perform the prediction.
upscaled_image = model.predict(input_image)
```

```python
import tensorflow as tf
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

input_image = tf.expand_dims(image, axis=0)  # Shape: (1, H, W, C)

# Predict
upscaled_image = model.predict(input_image)

# Set up the figure with GridSpec
fig = plt.figure(figsize=(10, 8))
gs = gridspec.GridSpec(2, 2)  # Make second row taller

# Plot original image
ax0 = fig.add_subplot(gs[0, 0])
ax0.imshow(image)
ax0.set_title("Original")
ax0.axis('off')

# Output keys and their subplot positions
output_keys = ['out_256', 'out_256_b', 'out_384']
positions = [gs[0, 1], gs[1, 0], gs[1, 1]]  # Place out_384 last and largest

for key, pos in zip(output_keys, positions):
    output = upscaled_image[key]
    output = np.squeeze(output, axis=0)
    output = np.clip(output, 0, 255).astype(np.uint8)

    img = Image.fromarray(output)
    img.save(f"{key}.png")

    ax = fig.add_subplot(pos)
    ax.imshow(img)
    ax.set_title(key)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

AI GENERATED BENCHMARK

```python
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import cv2


original_image = cv2.imread(orignalimage)
original_image_resized = cv2.resize(original_image, (256, 256))


for key in ['out_256', 'out_256_b', 'out_384']:
    generated_image = cv2.imread(f"{key}.png")
    generated_image_resized = cv2.resize(generated_image, (256, 256))

    psnr_value = psnr(original_image_resized, generated_image_resized)
    ssim_value = ssim(original_image_resized, generated_image_resized, multichannel=True, channel_axis=2)

    print(f"Metrics for {key}:")
    print(f"PSNR: {psnr_value}")
    print(f"SSIM: {ssim_value}")
    print("-" * 20)
```
